### Reading the classification report

`classification_report` prints one row per class, plus three summary rows:

```
              precision    recall  f1-score   support
      Repays       0.97      0.92      0.94      1800
     Default       0.50      0.70      0.58       200
    accuracy                           0.90      2000
   macro avg       0.73      0.81      0.76      2000
weighted avg       0.92      0.90      0.91      2000
```

- **support** — how many true examples of that class exist in the test set.
- **Per-class rows (Repays, Default)** — precision, recall, and F1
  computed treating that class as the positive class. Repays scores
  much higher because the model has 9× more training examples for it.
- **accuracy** — (TP+TN) / total across all classes. On imbalanced data
  this is misleading: a model predicting "repays" for everyone would
  also score 90% while catching zero defaulters.
- **macro avg** — plain average of the two class rows, treating both
  classes equally regardless of size. Use this when the minority class
  matters as much as the majority.
- **weighted avg** — average weighted by support. Repays contributes
  1800/2000 of the weight and dominates — hiding how poorly the model
  performs on the class that actually matters.

## Challenge 7 — Compute F1 and Understand the Harmonic Mean

The F1 score is the harmonic mean of precision and recall:

$$F1 = \frac{2 \cdot P \cdot R}{P + R}$$

The harmonic mean is pulled toward the *lower* value. A high F1
requires *both* precision and recall to be high — one strong number
cannot rescue a weak one.
