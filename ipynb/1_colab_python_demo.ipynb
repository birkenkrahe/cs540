# 2. Download CSV data from Kaggle into `DataFrame`

We will download the `Iris` dataset from Kaggle. We will store the URL in a variable, and then read the CSV data into a `DataFrame`.

```python
url = 'https://tinyurl.com/iris-data-csv' # save URL as a string
```

Now we load the dataset with the `read_csv()` method from `pandas` and store it in `df`:

```python
df = pd.read_csv(url)
print(df)
```

## What is the Iris Dataset?

The Iris dataset is a classic and widely used dataset in machine learning and statistics. It contains 150 samples of iris flowers, with each sample having four features (sepal length, sepal width, petal length, and petal width) and one of three possible species (setosa, versicolor, or virginica).

It is often used for introductory classification tasks due to its simplicity and clear separation of classes.

### Where does it come from?

The Iris dataset was introduced by the British statistician and biologist Ronald Fisher in 1936. While the original source is often cited as his paper, the data is widely available and often presented in educational contexts.

You can find the dataset, for example, on the UCI Machine Learning Repository:
[https://archive.ics.uci.edu/dataset/53/iris](https://archive.ics.uci.edu/dataset/53/iris)
